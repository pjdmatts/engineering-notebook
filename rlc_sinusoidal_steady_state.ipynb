```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ivsfreq(omega, vrms, R, L, C):
    xl = omega*L
    xc = 1/(omega*C)
    z = np.sqrt(R**2 + (xl - xc)**2)
    return vrms/z
```

```python
om = np.linspace(8e6, 12e6, 1000)
i_rms_2 = ivsfreq(om, 5e-3, 2, 5e-6, 2e-9)
i_rms_5 = ivsfreq(om, 5e-3, 5, 5e-6, 2e-9)
i_rms_10 = ivsfreq(om, 5e-3, 10, 5e-6, 2e-9)
```

```python
fig, ax = plt.subplots()
ax.plot(om/1e6, i_rms_2*1000, label= "R = 2$\:\Omega$")
ax.plot(om/1e6, i_rms_5*1000, label= "R = 5$\:\Omega$")
ax.plot(om/1e6, i_rms_10*1000, label= "R = 10$\:\Omega$")
ax.legend(loc= 'upper right')
plt.title("Current versus frequency, RLC. $V_{rms}\:=\:5.0mV, C\:=2.0\:nF, L\:=\:5.0\:\mu H$")
plt.xlabel("$\omega \: (x 10^{6} rad/s)$")
plt.ylabel("$I_{rms}$ (mA)")
plt.show()
fig.savefig('RLC_SteadyState.png')
fig.savefig('RLC_SteadyState.jpg')
fig.savefig('RLC_SteadyState.pdf')
```
